```python

#!pip install PyPortfolioOpt as pypfopt

import pandas as pd
import numpy as np
import random
from pandas_datareader import data as web
import matplotlib.pyplot as plt
from datetime import datetime
plt.style.use('fivethirtyeight')
from matplotlib.ticker import FuncFormatter

import scipy.optimize as sco
plt.style.use('fivethirtyeight')
np.random.seed(777)
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
tickers = ['AAPL','VOO', 'SCHD','QCLN','QQQ','VNQ','TSLA']
beginnign_date='15-09-2012'
```

```python
data = pd.DataFrame()
for item in tickers:
    data[item] = web.DataReader(item, data_source='yahoo', start=beginnign_date )['Adj Close']

data.head()
```

```python
title = 'Portfolio Adj. Close Price History    '
my_stocks = data
plt.figure(figsize=(12.2,4.5)) 
for c in my_stocks.columns.values:
  plt.plot( my_stocks[c],  label=c)
plt.title(title)
plt.xlabel('Date',fontsize=18)
plt.ylabel('Adj. Price USD ($)',fontsize=18)
plt.legend(my_stocks.columns.values, loc='upper left')
plt.show()
```

```python
daily_simple_returns = data.pct_change()
daily_simple_returns.plot()
```

```python
annual_returns = daily_simple_returns.mean() * 250
annual_returns
```

```python
current_prices = data.iloc[len(data.index)-1].to_numpy()
weights = [x / sum(current_prices) for x in current_prices]
weights
```

```python
returns = data.pct_change()
cov_matrix_annual = returns.cov() * 252
cov_matrix_annual
```

```python
port_variance = np.dot(weights, np.dot(cov_matrix_annual, weights))
port_variance
```

```python
port_volatility = np.sqrt(port_variance)
port_volatility
```

```python
portfolioSimpleAnnualReturn = np.sum(returns.mean()*weights) * 252
portfolioSimpleAnnualReturn
```

```python
percent_var = str(round(port_variance, 4) * 100) + '%'
percent_vols = str(round(port_volatility, 4) * 100) + '%'
percent_ret = str(round(portfolioSimpleAnnualReturn, 4)*100)+'%'
print("Expected annual return : "+ percent_ret)
print('Annual volatility/standard deviation/risk : '+percent_vols)
print('Annual variance : '+percent_var)
```
